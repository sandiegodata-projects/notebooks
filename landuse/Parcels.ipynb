## San Diego County Parcel Land Use

SANGIS publishes a geographic file that links basic tax assessment data to parcels in San Diego. The file appears to be comprehensive, with about 1M parcel records. The data in the file includes fields for things like:
* identification numbers
* Owner names and addresses
* Legal name, for businesses
* Several different types of land use and zoning codes
* Tax assessment value for land and impovements

While the file is regularly updated -- it appears to be updated approximately monthly -- the documentation notes:

> Parcel Zone and Land Use codes are used by County appraisers. They are 
> not well maintained and are not considered reliable for other purposes

There are other files that are more reliable for this information, but this file should be sufficient for "order of magnitude" queries about how value and land use are connected. 

This notebook uses a lot of codes, which are defined in the [documentation on the SANGIS website](http://www.sangis.org/download/index.html). 


```python
%matplotlib inline
import ambry
import pandas as pd
l = ambry.ilibrary()
p = l.get("sangis.org-parcels-orig-parcels-geo").partition
```

```python
df = p.select("SELECT own_name1, legldesc, asr_zone, asr_landus, asr_total, nucleus_zo, nucleus_us  FROM parcels").pandas
df['own_name1'].count()
```

## Land Use Codes and Assessed Value

The `nucleus_us` field, ( "Nucleaus Use" ) is a detailed use code, whe names for which are defined in  the [documentation on the SANGIS website](http://rdw.sandag.org/file_store/Parcel/Parcels.pdf). 

Here is a list of the codes, total assess value in dollars, and average assessed value for the top 10 categories by  number. The value sums and counts are for the whole county. The codes are:

* 069 INFO PARCEL - MH SPACE IN RENTAL PARK
* 060 INFO PARCEL-GENERIC

The `000` code is not defined in the documentation for the `nucleus_us` field.

```python
x = df[['nucleus_us','asr_total','asr_landus' ]].groupby(df.nucleus_us)
sums = x.sum().sort('asr_total', ascending=False)
counts = x.count()['asr_total']
df2 = pd.DataFrame(index=sums.index)
df2['sums'] = sums['asr_total']
df2['counts'] = counts
df2['avg'] = df2['sums'] / df2['counts']
df2.sort('counts', ascending=False).head(10)
```
